```python
import networkx as nx
def make_sub(G:nx.Graph,id,knn_result,dictionary_label):
    # print(G)
    l = len(knn_result)
    g = nx.Graph()
    for key,value in dictionary_label.items():
        if value == id:
            # print(key)
            g.add_node(key)
    l_nodes = list(g.nodes)
    # print(l_nodes)
    size = len(l_nodes)
    # print(type(l_nodes))
    for i in range (0,size):
        for j in range (i+1,size):
            # print(l_nodes[i],l_nodes[j])
            if G.has_edge(l_nodes[i],l_nodes[j]) == True:
                g.add_edge(l_nodes[i],l_nodes[j])
            else:
                continue
    return g

```
