```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
# Returns theta in [-pi/2, 3pi/2]
def generate_theta(a, b):
    u = random.random() / 4.0
    theta = np.arctan(b/a * np.tan(2*np.pi*u))

    v = random.random()
    if v < 0.25:
        return theta
    elif v < 0.5:
        return np.pi - theta
    elif v < 0.75:
        return np.pi + theta
    else:
        return -theta

def radius(a, b, theta):
    return a * b / np.sqrt((b*np.cos(theta))**2 + (a*np.sin(theta))**2)

def random_point(a, b, position_x,position_y):
    random_theta = generate_theta(a, b)
    max_radius = radius(a, b, random_theta)
    random_radius = max_radius * np.sqrt(random.random())

    return np.array([
        random_radius * np.cos(random_theta)+position_x,
        random_radius * np.sin(random_theta)+position_y
    ])




def point_on_triangle(pt1, pt2, pt3):
    """
    Random point on the triangle with vertices pt1, pt2 and pt3.
    """
    x, y = sorted([random.random(), random.random()])
    s, t, u = x, y - x, 1 - y
    return (s * pt1[0] + t * pt2[0] + u * pt3[0],
            s * pt1[1] + t * pt2[1] + u * pt3[1])

pt1 = (0, 16)
pt2 = (0, 0)
pt3 = (16, 16)
points = [point_on_triangle(pt1, pt2, pt3) for _ in range(250)]

pt4 = (0, 0)
pt5 = (16, 0)
pt6 = (16, 16)
points2 = [point_on_triangle(pt4, pt5, pt6) for _ in range(250)]
import matplotlib.pyplot as plt
x1,y1 = zip(*points2)
x2,y2 = zip(*points)

points = np.array([random_point(2,1,10,15) for _ in range(1500)])
points2 = np.array([random_point(4, 5,5,5) for _ in range(1000)])
points3 = np.array([random_point(2, 1,5,15) for _ in range(1500)])
points4 = np.array([random_point(1, 1,15,10) for _ in range(500)])
points5 = np.array([random_point(1, 1,15,7) for _ in range(500)])


xi = np.concatenate((points[:,0] , points2[:,0],points3[:,0],points4[:,0],points5[:,0],x1,x2),axis=0)
print(type(xi))
yi = np.concatenate((points[:,1] , points2[:,1],points3[:,1],points4[:,1],points5[:,1],y1,y2),axis=0)
print(type(yi))

dfx = pd.DataFrame(xi)
dfy = pd.DataFrame(yi)
df = pd.DataFrame()
df['x'] = dfx[0]
df['y'] = dfy[0]
print(df)
df.to_csv('./sample1.csv')
plt.scatter(df['x'], df['y'],s=1)
plt.show()
```
